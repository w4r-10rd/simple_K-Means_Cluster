```python
import numpy as np
from sklearn.cluster import MiniBatchKMeans
from sklearn.datasets import make_blobs
```

```python
# Generate synthetic dataset
X, _ = make_blobs(n_samples=10000, n_features=20, centers=10, random_state=42)
```

```python
# Initialize heavy k-means model
kmeans = MiniBatchKMeans(n_clusters=10, batch_size=100, max_iter=100)

# Train the model
kmeans.fit(X)
```

```python
# Get cluster centroids
centroids = kmeans.cluster_centers_

# Get cluster labels for each data point
labels = kmeans.labels_

# Evaluate the model
inertia = kmeans.inertia_

print("Cluster centroids:\n", centroids)
print("Cluster labels:\n", labels)
print("Inertia (Within-cluster sum of squares):", inertia)
```

```python
import matplotlib.pyplot as plt

# Plot data points with cluster assignments
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=20, alpha=0.5)
plt.scatter(centroids[:, 0], centroids[:, 1], marker='o', c='red', s=100, label='Centroids')
plt.title('Clustering Results')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.grid(True)
plt.show()
```
